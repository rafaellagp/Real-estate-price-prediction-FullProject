```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
colors = sns.color_palette('mako')[2:6]
cmap = 'mako' 
```

```python
df = pd.read_csv("clean_data.csv")
df.head()
```

```python
#target
x = df.price
sns.histplot(x)
plt.show()
```

```python
sns.boxplot(x=df["price"])
```

```python
corr = df.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
corr
```

```python
plt.subplots(figsize=(16, 8))   
sns.heatmap(df.corr(), annot=True, linewidths=0.05, fmt= '.2f', vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(20, 220, n=200))
plt.savefig('corr-matrix.png')
plt.show()
```

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots(1,1, figsize=(8, 5), dpi=80)
my_order = df.groupby(by=["property_type"])["price"].mean().sort_values(ascending=False).index
box_plot = sns.boxplot(x = 'property_type',y = 'price',data = df, order=my_order, showfliers=False, palette="mako")
plt.ylabel('Price', fontsize=15, fontweight='bold', fontfamily='times')
plt.xlabel('')
ax = box_plot.axes
ax.text(x=0.5, y=1.1, s='Comparison of real estate prices of regarding the property type in Belgium', fontfamily='times', fontsize=15, weight='bold', ha='center', va='bottom', transform=ax.transAxes)
ax.text(x=0.5, y=1.05, s='source: immoweb.com', fontsize=10, alpha=0.75, ha='center', fontfamily='times', va='bottom', transform=ax.transAxes)
lines = ax.get_lines()
ax.tick_params(axis='x', rotation=90)
for s in ['top', 'left', 'right']:
    ax.spines[s].set_visible(False)
ax.grid(axis='y', linestyle='-', alpha=0.4) 

box_plot.figure.tight_layout()
```
